```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("studentscores.csv.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

DROP UNNAMED COLUMNS:

```python
df = df.drop('Unnamed: 0', axis = 1)
print(df.head())
```

CHANGE WEEKLY HOUR COLOUMNS:

```python
df ['WklyStudyHours'] = df ['WklyStudyHours'].str.replace('05-Oct','5-10')
print(df.head())
```

GENDER DISTRIBUTION:

```python
plt.Figure(figsize = (5,5))
ax = sns.countplot(data = df, x = 'Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

"FROM THE ABOVE DATA I HAVE ANALYZE FEMALES MORE THAN MALES"

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean','ReadingScore': 'mean', 'WritingScore': 'mean' })
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between the Studends and their Parents Education")
plt.show
```

"WE HAVE ANALYZED EDUCATION OF PARENT HAS INFLUENCED ON STUDENT'S SCORE"

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean','ReadingScore': 'mean', 'WritingScore': 'mean' })
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relation ship between Parents and their Students")
plt.show
```

"I HAVE ANALYZED FROM THE ABOVE CHART THERE IS NO SPECIAL IMPACT ON STUDENT'S BEHALF OF PARENT MARTIAL STATUS"

"DISTRIBUTION OF ETHINIC GROUP"

```python
group_A = df.loc[(df['EthnicGroup'] == "group A")].count()
group_B = df.loc[(df['EthnicGroup'] == "group B")].count()
group_C = df.loc[(df['EthnicGroup'] == "group C")].count()
group_D = df.loc[(df['EthnicGroup'] == "group D")].count()
group_E = df.loc[(df['EthnicGroup'] == "group E")].count()


l = ['GroupA', 'GroupB', 'GroupC', 'GroupD', 'GroupE']
my_list = [group_A['EthnicGroup'], group_B['EthnicGroup'], group_C['EthnicGroup'], group_D['EthnicGroup'], group_E['EthnicGroup']]
plt.pie(my_list, labels = l, autopct = '%1.2f%%')
plt.title("Distribution of Ethnic Group")
plt.show()
```
